```python
#DOWNLOAD AND INSTALL THE NECESSARY LIBRARIES & MODULES:
#!pip install -U sapcy
#!python -m spacy download en_core_web_sm
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
#variable to store Stop Words
stopWords = list(STOP_WORDS)
stopWords
```

```python
#objects to store formatted document
nlp = spacy.load('en_core_web_sm')
nlp
```

```python
#Variable to store formatted document
doc = nlp(TEXT)
doc
```

```python
#WORD-TOKENIZATION: Variable to sore tokens as word
tokens = [token.text for token in doc]
print(tokens)
```

```python
punctuation
```

```python
#Variable to store punctuation  & escape sequence- next line.
punctuation = punctuation + '\n'
punctuation
```

```python
#Dictionary to store word count:
word_count = {}
for word in doc:
    if word.text.lower() not in stopWords:
        if word.text.lower() not in punctuation:
            if word.text not in word_count.keys():
                word_count[word.text] = 1
            else:
                word_count[word.text] = word_count[word.text]+1
word_count
```

```python
#Variable to store max count:
Max_count = max(word_count.values())
Max_count
```

```python
for word in word_count.keys():
    word_count[word] = word_count[word]/Max_count
word_count    
```

```python
#SENTENCE TOKENIZATION: Variable to store sentences as tokens 
sent_tokens = [sent for sent in doc.sents]
sent_tokens
```

```python
#Dictionary to store sentence score
sent_scores = {}
for sent in sent_tokens:
    for word in sent:
        if word.text.lower() in word_count.keys():
            if sent not in sent_scores.keys():
                sent_scores[sent] = word_count[word.text.lower()]
            else:
                sent_scores[sent] = sent_scores[sent]+word_count[word.text.lower()]
sent_scores
```

```python
# From heapq module import the nlargest function: NO.OF SENT
from heapq import nlargest
select_len = int(len(sent_tokens)*0.3)
select_len
```

```python
#Get the n sentences with top scores:
summary = nlargest(select_len, sent_scores, key = sent_scores.get)
summary
```

```python
#Marking every value of the list:
summary = [word.text for word in summary]
summary
```

```python
#Combining the sentences excluding extra spaces:
final_summary = ' '.join(summary)
final_summary
```
